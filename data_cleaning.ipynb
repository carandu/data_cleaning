```python
import numpy as np
import pandas as pd
import re
```

```python
df = pd.read_csv('/Users/carandu/thinkful_projects/data_cleaning_challenge/WELLCOME_APCspend2013_forThinkful.csv', encoding='latin1')
```

```python
df.head()
```

```python
df.columns = ['PMID/PMCID', 'Publisher', 'Journal', 'Article', 'Cost']
df.dtypes
```

```python
df['Cost'] = df['Cost'].str.replace('£', '')
df['Cost'] = df['Cost'].str.replace('$', '')
df['Cost'] = df['Cost'].astype(float)
df.head(200)
```

```python
acs = df['Publisher']
american_chemical_society = acs.str.contains('Amercian Chemical Society', 'Amercian Chemical Society')
american_chemical_society[:24]
```

```python
# de-dupe the id column
# first we find the PMIDs and PMCIDs and separate these into two columns
# this function obtains the 7-digit id that starts with 3 from the alphanumeric id
# col, which is the PMCID
def get_pmcid(x):
    pmcid = re.findall('3\d{6}', str(x))
    if len(pmcid) > 0:
        return pmcid[- 1]

# this function obtains the 8-digit id that starts with 2 from the alphanumeric id
# col, which is the PMID
def get_pmid(x):
    pmid = re.findall('2\d{7}', str(x))
    if len(pmid) > 0:
        return pmid[0]
```

```python
df['PMID'] = df['PMID/PMCID'].apply(get_pmid)
df['PMCID'] = df['PMID/PMCID'].apply(get_pmcid)
# now drop the old column
del df['PMID/PMCID']
# drop rows with duplicate ids (always change in place in case you have large dfs):
df.drop_duplicates(subset='PMID', keep="last", inplace=True)
df.drop_duplicates(subset='PMCID', keep="last", inplace=True)
# drop rows with no IDs
df.drop(df[(df.PMID == None) & (df.PMCID == None)].index, inplace=True)
```
